```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import math as m
```

```python
size=0.035
path = f'rotation/Circle_{size}/data360/circle_{size}_0_degrees.csv'
df = pd.read_csv(path)
```

```python
df
```

```python
# https://isopromat.ru/teormeh/kratkaja-teoria/kinematika-skorost-tocki
def get_velocity(timestamp):
    if timestamp == 0:
        return 0
    else:
        dx = (df.loc[timestamp]['x'] - df.loc[timestamp - 1]['x']) / 0.001
        dy = (df.loc[timestamp]['y'] - df.loc[timestamp - 1]['y']) / 0.001
        dz = (df.loc[timestamp]['z'] - df.loc[timestamp - 1]['z']) / 0.001
        return np.sqrt(dx ** 2 + dy ** 2 + dz ** 2) # m/milisecs
  
```

```python
# add velocity column to df
velocities = np.empty((1000,))
for i in range(1000):
    velocities[i] = get_velocity(i)
df['velocity'] = velocities
```

```python
df
```

```python
def get_acceleration(timestamp):
    if timestamp == 0:
        return 0
    else:
        a = (df.loc[timestamp]['velocity'] - df.loc[timestamp - 1]['velocity']) / 0.001
        return a 
```

```python
# add velocity column to df
accelerations = np.empty((1000,))
for i in range(1000):
    accelerations[i] = get_acceleration(i)
df['acceleration'] = accelerations
```
